Pattern

```python
import time, sys
```

```python
indent = 0
indentIncresing = True
```

```python
try:
    while True:
        print(' ' * indent, end='')
        print('********')
        time.sleep(0.1)
        if indentIncresing:
            indent += 1
            if indent == 20:
                indentIncresing
        else:
            indent -= 1
            if indent == 0:
                indentIncresing = True
except KeyboardInterrupt:
    sys.exit()
```
